## Matplotlib Part 2 - how to create subplots using OOP

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x, y)
```

```python
axes[0].plot(x, y)
axes[0].set_title('Line Plot')
axes[1].plot(y, x)
axes[1].set_title('Line Plot - Reverse')
```

## MATPLOTLIB PART 3 - SETTING COLORS, LINEWIDTH, LINE TYPES

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='#FF0000', linewidth=1.0, label='line')
ax.set_xlabel('X AXIS', fontsize=12)
ax.set_ylabel('Y AXIS', fontsize=12)
ax.set_title('A Simple Plot', fontsize=14)

ax.legend(loc='upper left')

```

```python
# plot appearance
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, 
        color='#FF0000', 
        linewidth=2, 
        linestyle=':', 
        label='x', 
        marker="o", 
        markersize=10,
        markerfacecolor='#000000',)
```

## EXERCISES

```python
import numpy as np

x = np.arange(0, 100)
y = x * 2
z = x ** 2
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
fig = plt.figure()

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title(' Simple Figure Object')
```

```python
fig = plt.figure(figsize=(8, 6))

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x, y, color='r', linewidth=2)
ax2.plot(x, y, color='r', linewidth=2)

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([.2, .5, .4, .4])
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([.2, .5, .4, .4])

ax.plot(x, z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

ax2.plot(x, y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim([20, 22])
ax2.set_ylim([30, 50])

```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(x, y, 'b', linewidth=3, ls='--')
axes[1].plot(x, z, 'r-', lw=3, ls='--')
```
