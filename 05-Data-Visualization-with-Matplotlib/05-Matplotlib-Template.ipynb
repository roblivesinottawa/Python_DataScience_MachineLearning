```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot')

plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, 'r')
```

```python
plt.subplot(1,2,2)
plt.plot(y, x, 'b')
```

```python
# object oriented method
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')


```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

ax1.plot(x, y)
ax1.set_title('LARGER_PLOT')

ax2.plot(y, x)
ax2.set_title('SMALLER_PLOT')
```

## Matplotlib Part 2 - how to create subplots using OOP

```python
fig, ax = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

for current_ax in ax:
    current_ax.plot(x, y)
```
